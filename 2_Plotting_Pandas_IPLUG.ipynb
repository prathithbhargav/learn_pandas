# 2b Let's try and see if we can plot two things at the same time. 

```python
df.plot(x='Date',y=['dailyconfirmed','dailydeceased'])
```

Now let's normalise both dailyconfirmed and dailydeceased

```python
df['dailyconfirmed'] = df['dailyconfirmed']/df['dailyconfirmed'].max()

### do the same for dailydeceased
df['dailydeceased'] = df['dailydeceased']/df['dailydeceased'].max()
```

```python
df.plot(x='Date',y=['dailyconfirmed','dailydeceased'],rot=90)
```
